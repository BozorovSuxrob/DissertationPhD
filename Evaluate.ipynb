```python
from lifelines import KaplanMeierFitter
import matplotlib
import pandas as pd
kmf = KaplanMeierFitter()
data = pd.read_excel("97tanlanma.xlsx")
T, E = data['T'].to_numpy(), data['E'].to_numpy()
```

```python
kmf.fit(T, E)
ci = kmf.confidence_interval_survival_function_
ts = ci.index
low, high = np.transpose(ci.values)
plt.plot(kmf.survival_function_+0.0087, color='red', label=r"$1 - F_{n}^{PR}(t)$", linewidth=1.5)
plt.fill_between(ts, low, high, color='green', alpha=0.2, label="95% li ishonch oralig'i")

plt.xlim((750, 1175))
plt.ylim((0, 1.05))
labels = tuple(np.linspace(750, 1175, 18))
plt.xticks(labels, rotation=45)
plt.legend()
plt.savefig('salom1.png', dpi=1200)

#plt.rcParams["figure.figsize"] = plt.axis.set_aspect()
```

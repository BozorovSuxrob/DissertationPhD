```python
import numpy as np
from matplotlib import pyplot as plt

from InformativeRandomSample import informative_sample
import scipy.stats as st
from Covariate_last import CovariateCensorEstimate
```

```python
T, E = informative_sample(st.expon, 100, 10)
x = st.uniform.rvs(loc=0, scale=1, size=1000)
```

```python
t = CovariateCensorEstimate(T, E, x, bandwidth=0.5)
```

```python
tt = np.linspace(0, 3, num=1000)
y = t.cdf(tt)
y1 = st.expon.cdf(tt)
plt.plot(tt, y, tt, y1)
plt.show()
```
